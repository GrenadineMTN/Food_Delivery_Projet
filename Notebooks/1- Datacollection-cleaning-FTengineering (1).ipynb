Import des librairies : 

```python
import numpy as np
import pandas as pd
from geopy.distance import geodesic
```

1- Chargement des datasets train, test, et concaténation avec le dataset Sample Submission : 

```python
#Chargement et concaténation des datasets train et test : 
train=pd.read_csv('Food Delivery/train.csv')
test=pd.read_csv('Food Delivery/test.csv')
df_initial=pd.concat([train,test]).reset_index(drop=True)
df_initial
```

```python
#Remplacement du type de Time_taken en float pour homogénéiser avec le dataset sample
df_initial['Time_taken(min)']=df_initial['Time_taken(min)'].str.replace('(min)','').astype(float)
```

```python
#Lecture du fichier sample avec un merge dans le dataset 
sample=pd.read_csv('Food Delivery/Sample_Submission.csv')
df=pd.merge(df_initial,sample, how='left', on='ID')
df
```

```python
#Remplacement des NA par les valeurs de Sample
df['Time_taken(min)'].fillna(df['Time_taken (min)'],inplace=True)
```

```python
df=df.drop(columns=['Time_taken (min)'])
```

```python
df
```

2- Data Cleaning : 

```python
#Remplacer conditions dans la colonne "Weatherconditions" par rien
df['Weatherconditions']=df['Weatherconditions'].str.replace('conditions','')
```

```python
#Constatation de l'absence de vraies valeurs NAN dans le dataset:
df.isnull().sum()
```

```python
#Remplacement des fausses valeurs NaN par des vrais NaN
df.replace('NaN', np.nan, inplace=True)
df.replace('NaN ',np.nan, inplace=True)
df.replace(' NaN',np.nan, inplace=True)
df.replace('nan',np.nan, inplace=True)
df.replace('nan ',np.nan, inplace=True)
```

```python
#Détection des valeurs NaN dans le Dataset
df.isnull().sum()
```

```python
#Spliter city_code de Delivery_person_ID
df['City_code']=df['Delivery_person_ID'].str.split('RES', expand=True)[0]
```

```python
#Supprimer les espaces au début et à la fin des chaînes de caractères
df=df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#Vérification des valeurs dupliquées
df.duplicated().value_counts()
```

```python
#Modification du type de certaines colonnes
def update_datatype(df):
    df['Delivery_person_Age'] = df['Delivery_person_Age'].astype('float64')
    df['Delivery_person_Ratings']=df['Delivery_person_Ratings'].astype('float64')
    df['multiple_deliveries']=df['multiple_deliveries'].astype('float64')
    df['Order_Date']=pd.to_datetime(df['Order_Date'], format="%d-%m-%Y")

update_datatype(df)
    
```

```python
#Imputation des valeurs NaN
def handle_null_values(df):
    df['Delivery_person_Age'].fillna(np.random.choice(df['Delivery_person_Age']),inplace=True)
    df['Delivery_person_Ratings'].fillna(df['Delivery_person_Ratings'].median(),inplace=True)
    df['Weatherconditions'].fillna(np.random.choice(df['Weatherconditions']),inplace=True)
    df['Road_traffic_density'].fillna(df['Road_traffic_density'].mode()[0], inplace=True)
    df['multiple_deliveries'].fillna(df['multiple_deliveries'].mode()[0],inplace=True)
    df['Festival'].fillna(df['Festival'].mode()[0],inplace=True)
    df['City'].fillna(df['City'].mode()[0],inplace=True)
handle_null_values(df)
```

```python
#Revérification des valeurs nulles dans le Dataset
df.isna().sum()
```

3- Feature Engineering: 

```python
#Convertir time ordered et time order picked en date time
df['Time_Orderd']=pd.to_datetime(df['Time_Orderd'])
df['Time_Order_picked']=pd.to_datetime(df['Time_Order_picked'])
```

```python
#Calculer la différence entre les deux 
df['Time_Preparation'] = df['Time_Order_picked'] - df['Time_Orderd']
#Convertir la différence en minutes seulement Preparation
df['Time_Preparation']=df['Time_Preparation'].dt.seconds//60
#Imputer les valeurs NaN de Time Preparation par la médiane
df['Time_Preparation'].fillna(df['Time_Preparation'].median(),inplace=True)
```

```python
#Suppression de la colonne  Time Orderd 
df=df.drop(columns='Time_Orderd')
```

```python
df.isna().sum()
```

```python
#Calcul de la distance entre la localisation du restaurant et celle du client:
from geopy.distance import geodesic
def calculate_distance(df):
    df['distance']=np.zeros(len(df))
    restaurant_coordinates=df[['Restaurant_latitude','Restaurant_longitude']].to_numpy()
    delivery_location_coordinates=df[['Delivery_location_latitude','Delivery_location_longitude']].to_numpy()
    df['distance']=np.array([geodesic(restaurant, delivery) for restaurant, delivery in zip (restaurant_coordinates, delivery_location_coordinates)])

calculate_distance(df)
```

```python
#Transformation de la distance en chaine de caractères pour éviter des erreurs dans la conversion
df['distance']=df['distance'].astype("str")
#Remplacement de km par rien 
df['distance']=df['distance'].str.replace(' km','').astype(float)
#Limitation des nombres après la virgule à 2 
df['distance']=round(df['distance'],2)
```

```python
#Suppression des distances supérieures à 50km
df = df[~(df["distance"] > 50)]\
    .reset_index(drop=True).copy()
```

```python
#Extraction des dates features de order date 
def extract_date_features (data):
  data["day"] = data.Order_Date.dt.day
  data["month"]=data.Order_Date.dt.month
  data['day_of_week']=data.Order_Date.dt.day_of_week.astype(int)
  data['is_weekend']=np.where(data['day_of_week'].isin([5,6]),1,0)

extract_date_features(df)

#Extraction des heures de time order picked
df['Time_Order_picked']=df['Time_Order_picked'].dt.hour


#Définir plage horaire de time order picked
time_ranges = [(0, 5), (6, 11), (12, 17), (18, 23)]
time_categories = ['night', 'morning', 'afternoon', 'evening']

# Création d'une fonction lambda pour la définition des plages horaires
get_time_category = lambda hour: next((category for (start, end), category in zip(time_ranges, time_categories) if start <= hour <= end), 'unknown')

# Application de la fonction lambda à la colonne 'Time_category'
df['Time_category'] = df['Time_Order_picked'].apply(get_time_category)
```

```python
#Renommer colonne Time_order_picked 
df.rename(columns={'Time_Order_picked': 'Hour_picked'},inplace=True)
```

```python
#Export du dataset sous format csv 
df.to_csv('df_final.csv')
```
